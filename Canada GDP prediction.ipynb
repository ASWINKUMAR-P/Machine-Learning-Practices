```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("canada.csv")
```

```python
df.head()
```

```python
import seaborn as sns
```

```python
sns.regplot(x='year',y='per capita income (US$)',data=df)
```

```python
x=df.drop('per capita income (US$)',axis=1)
```

```python
y=df.drop('year',axis=1)
```

```python
brain = LinearRegression()
```

```python
brain.fit(x,y)
```

```python
brain.score(x,y)
```

```python
brain.predict([[3000000010]])
```
